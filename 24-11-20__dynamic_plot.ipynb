```python
from utils import GraphEdgeIterator, generate_seeds, iteration_logging
from tqdm.notebook import tqdm
import sys
```

```python
edges = list(GraphEdgeIterator())
```

```python
for buffer_count in (2, 5):
    list(tqdm(GraphEdgeIterator(buffer_count=buffer_count), total=len(edges), desc=f"Edges {buffer_count}"))
```

```python
wrapped_edges = tqdm(GraphEdgeIterator(), total=len(edges), desc="Edges")
```

```python
seeds = list(generate_seeds(wrapped_edges, omega=10))
```

```python
from sympy import Interval

roots = {}
for s, (ts, te), candidates in seeds:
    roots.setdefault(s, {})
    roots[s][Interval(ts, te)] = candidates
```

```python
for root, intervals in roots.items():
    if len(intervals) > 1:
        print(root)
        for interval, candidates in intervals.items():
            print(interval, candidates)
```
