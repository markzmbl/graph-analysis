```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for omega in (10, ):
    stresstest_df = pd.read_csv(f"june-run-omega00{omega}_memory.csv").set_index("time_seconds")
    stresstest_df.index -= stresstest_df.index.min()
    stresstest_df.index /= 3600
```

```python
feature = (stresstest_df["iterations_total"] / stresstest_df.index).loc[1:]
plt.plot(feature)
```

```python
feature = (stresstest_df["cycles_total"] / stresstest_df.index).loc[1:]
plt.plot(feature)
```
